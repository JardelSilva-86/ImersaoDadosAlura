1. Realizar a média da metragem para cada um dos bairros. (Paulo)

```python
# Não consegui resolver

# Solução Eduardo Iris Silva

# nome dos bairros
nome_bairro = df['Bairro'].values
dict_media = {} # criando um dicionário vazio

for bairro in nome_bairro:
    media_metragem = df.loc[df['Bairro'] == bairro, 'Metragem'].mean() # calcula a média da metragem de cada bairro
    dict_media[bairro] = media_metragem # Adiciona a média ao dicionário

dict_media
```

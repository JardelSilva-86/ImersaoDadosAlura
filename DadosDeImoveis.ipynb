```python
# diminuir as casa decimais, 2 formas para fazer isso.
pd.set_option('display.precision', 2) # 1º forma
pd.set_option('display.float_format', lambda x: '%.2f' % x) # 2º forma
dados_filtrados.describe()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set()
plt.figure(figsize=(10, 8))
ax = sns.histplot( data = dados_filtrados, x= 'Valor_anuncio_float', kde= True);
ax.set_title('Histograma Valor Imóvel')
plt.xlim((-50, 10000000))
plt.show()
```
